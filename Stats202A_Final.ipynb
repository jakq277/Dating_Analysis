```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/content/gdrive/MyDrive/202data/Speed_Dating_Data.csv", encoding = "latin")
college_data = data[data["wave"] == 5]
```

```python
college_data
```

```python
college_data = college_data[["gender","field_cd","race_o","age_o","date"]].dropna()
```

```python
X = college_data[["gender","field_cd","race_o","age_o"]].applymap(str)
y = college_data["date"]
```

```python
X.columns
```

```python
X[['gender', 'field_cd', 'race_o',]]
```

```python
one_hot = pd.get_dummies(X[['gender', 'field_cd', 'race_o',]])
```

```python
one_hot
```

```python
X = pd.DataFrame(college_data["age_o"]).join(one_hot)
```

```python
X
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=123,test_size=0.2)

lm = LinearRegression()
lm.fit(X_train,y_train)
y_pred = lm.predict(X_test)

print(f"mse of model on train_data is {mean_squared_error(lm.predict(X_train),y_train)}")
print(f"mse of model on test_data is {mean_squared_error(y_pred,y_test)}")
```

```python
from sklearn.neural_network import MLPRegressor

X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=123,test_size=0.2)

non_lm = MLPRegressor(random_state = 123)
non_lm.fit(X_train,y_train)
y_pred = non_lm.predict(X_test)
print(f"mse of model on train_data is {mean_squared_error(non_lm.predict(X_train),y_train)}")
print(f"mse of model on test_data is {mean_squared_error(y_pred,y_test)}")
```

```python
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest

select = SelectKBest(score_func=chi2, k=5)
selected_X = select.fit_transform(X,y)


X_train, X_test, y_train, y_test = train_test_split(selected_X, y,random_state=123,test_size=0.2)
non_lm = MLPRegressor(random_state = 123)
non_lm.fit(X_train,y_train)
y_pred = non_lm.predict(X_test)
print(f"mse of model on train_data is {mean_squared_error(non_lm.predict(X_train),y_train)}")
print(f"mse of model on test_data is {mean_squared_error(y_pred,y_test)}")
```
